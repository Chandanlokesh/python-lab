```python
#Implement a python program to demonstrate the following using NumPy a) Array manipulation, Searching, Sorting and splitting. b) broadcasting and Plotting NumPy arrays

import numpy as np
import matplotlib.pyplot as plt

arr = np.array([10,50,40,30,20,60])
print(f'original array {arr}')

reshape_arr = arr.reshape(2,3)
print(f'reshaped array {reshape_arr}')

index = np.where(arr==30)
print(f'index of 30 is {index}')

sort_arr = np.sort(arr)
print(f'sorted array {sort_arr}')

split_arr = np.array_split(arr,2)
print("Split array:", split_arr)

a = np.array([1,2,3])
b = np.array([4,5,6])

c = a+b
print("Broadcasted sum:", c)

x = np.arange(1,11)
y = x*x

plt.title('line graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y,color = 'red')
plt.show()
```
