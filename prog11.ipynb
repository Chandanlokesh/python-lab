```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate sample dataset
df = pd.DataFrame(np.random.rand(20, 2), columns=['Customer A', 'Customer B'])

# 1. Line Plot
print("Line Plot")
sns.lineplot(x='Customer A', y='Customer B', data=df)
plt.show()

# 2. Scatter Plot
print("Scatter Plot")
sns.scatterplot(x='Customer A', y='Customer B', data=df)
plt.show()

# 3. Bar Plot (showing each Customer A vs B pair)
print("Bar Plot")
sns.barplot(x='Customer A', y='Customer B', data=df)
plt.xticks(rotation=90)  # Rotate x-axis labels
plt.show()

# 4. KDE Plot (2D distribution for both columns)
print("KDE Plot")
sns.kdeplot(x='Customer A', y='Customer B', data=df)
plt.show()

```
