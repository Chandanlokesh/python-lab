```python
#write a python program to demonstrate the time series analysis with pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate a date range
date_range = pd.date_range(start='2024-01-01', periods=100, freq='D')
print("Date Range:")
print(date_range)

# Create a synthetic time series dataset
data = {
    'Date': date_range,
    'Value': np.cumsum(np.random.randn(100))  # Cumulative sum of random values
}
df = pd.DataFrame(data)
print("\nTime Series DataFrame:")
print(df)

# Set 'Date' as the index
df.set_index('Date', inplace=True)

# Display the first few rows of the dataset
print("\nFirst 5 rows of the dataset:")
print(df.head())

# Plot the time series
df.plot(figsize=(10, 5), title='Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid()
plt.tight_layout()
plt.show()

# Perform basic statistics
print("\nBasic Statistics:")
print(df.describe())

# Calculate rolling mean with a 7-day window
df.loc[:, 'Rolling Mean'] = df['Value'].rolling(window=7).mean()

# Plot the original data and rolling mean
df[['Value', 'Rolling Mean']].plot(figsize=(10, 5), title='Time Series with Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid()
plt.tight_layout()
plt.show()

# Resampling the data by month and calculating the monthly mean
df_resampled = df.resample('M').mean()
print("\nResampled Monthly Mean:")
print(df_resampled)

# Optional: Plot the resampled data
df_resampled.plot(figsize=(10, 5), title='Monthly Resampled Mean')
plt.xlabel('Month')
plt.ylabel('Average Value')
plt.grid()
plt.tight_layout()
plt.show()
```
