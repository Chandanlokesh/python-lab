```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 2x4 array (2 rows, 4 columns)
arr = np.array([(25, 67, 70, 30), (77, 99, 44, 60)])
print("Array:")
print(arr)

# Scatter plot
print("Scatter plot")
fig = plt.figure(figsize=(10, 5))  # Corrected 'figure' to 'figsize'

x = arr[0]  # First row
y = arr[1]  # Second row

plt.scatter(x, y, color='blue')
plt.xlabel('x-values')  # Fixed typo: xlable -> xlabel
plt.ylabel('y-values')  # Fixed typo: ylable -> ylabel
plt.title('Scatter plot of given data')
plt.show()
```
