```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create some sample time series data
data = {
    'Date': pd.date_range(start='2025-04-01', periods=10, freq='D'),  # 10 days
    'Sales': [50, 55, 45, 60, 52, 58, 65, 70, 68, 75]  # Daily sales
}

df = pd.DataFrame(data)

# Step 2: Set 'Date' as the index (important for time series)
df.set_index('Date', inplace=True)

# Step 3: Show the data
print("Time Series Data:\n", df)

# Step 4: Plot the time series
plt.plot(df.index, df['Sales'], marker='o', linestyle='-', color='orange')
plt.title('Lemonade Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

# Step 5: Do basic analysis
print("\nBasic Stats:")
print("Mean sales:", df['Sales'].mean())
print("Max sales on:\n", df[df['Sales'] == df['Sales'].max()])
print("Min sales on:\n", df[df['Sales'] == df['Sales'].min()])
```
