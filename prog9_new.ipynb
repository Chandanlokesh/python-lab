```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# 🍋 Step 1: Data - Temperature vs. Bought Lemonade (0 or 1)
X = np.array([[20], [22], [25], [27], [30]])  # Temperature
y = np.array([0, 0, 1, 1, 1])                 # Bought Lemonade? 0 = No, 1 = Yes

# 🧠 Step 2: Create a logistic regression model
model = LogisticRegression()

# 🏋️‍♂️ Step 3: Train the model
model.fit(X, y)

# 🔮 Step 4: Make a prediction - Will someone buy lemonade at 24°C?
temp = np.array([[24]])
prediction = model.predict(temp)
probability = model.predict_proba(temp)

print("Prediction (0=No, 1=Yes):", prediction[0])
print("Probability of buying lemonade:", probability[0][1])

# 🧾 Step 5: Print model info
print("Model Coefficients (m):", model.coef_)
print("Model Intercept (c):", model.intercept_)
import matplotlib.pyplot as plt

# Plot the sigmoid curve
temps = np.linspace(15, 35, 200).reshape(-1, 1)
probabilities = model.predict_proba(temps)[:, 1]

plt.plot(temps, probabilities, color='green')
plt.title("Logistic Regression: Probability of Buying Lemonade")
plt.xlabel("Temperature (°C)")
plt.ylabel("Probability of Purchase")
plt.grid()
plt.show()
```
