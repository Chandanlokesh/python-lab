```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
data = np.random.rand(10,10)

#line graph
plt.figure()
plt.plot(x, y, label='Sine')
plt.plot(x, z, label='Cosine')
plt.title('Line Graph')
plt.legend()

# Scatter Plot
plt.figure()
plt.scatter(x, y, c='red')
plt.title('Scatter Plot')

# Bar Graph
plt.figure()
plt.bar(['A', 'B', 'C', 'D', 'E'], [25, 30, 15, 20, 10])
plt.title('Bar Graph')

# Heatmap
plt.figure()
plt.imshow(data, cmap='viridis')
plt.title('Heatmap')
plt.colorbar()

plt.show()
```
